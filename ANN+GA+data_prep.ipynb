```
#utils
import math
def dist(y):
  d=math.sqrt((y[0][0]-y[1][0])**2+(y[0][1]-y[1][1])**2)
  return d
```

```
#utils
import itertools
def overlap(x):
  p1=[x[0],x[1]]
  p2=[x[2],x[3]]
  p3=[x[4],x[5]]
  p4=[x[6],x[7]]
  p5=[x[8],x[9]]
  p6=[x[10],x[11]]
  p=[p1,p2,p3,p4,p5,p6]
  comb=list(itertools.combinations(p, 2)) 
  c=[]
  for i in comb:
    c.append(dist(i))  
  ver=0
  for j in c:
    if j < 14.71:
      ver=ver+1
    else:
      pass  
  if ver>0:
    return False
  else:
    return True      
```

```
#utils
def weighter(x):
  r=[]
  lp=[]
  for i in x:
    i=i.tolist()
    r.append(i)
  #print(cost)  
  #print(r)
  for i in r:
    for j in i:
      for k in j:
        lp.append(-0.3*k+14)
  return lp
```

```
#utils
def prob(y):
  k=[]
  s=sum(y)
  for i in y:
    i=i/s
    k.append(i)
  return k
```

```
#utils
def mutation(p,mut_rate):
  h=[]
  for i in p:
    g=[]
    for t in i:
      if rd.uniform(0,1)<mut_rate:
        t=round(rd.uniform(5,37),2)
        g.append(t)
      else:
        g.append(t)
    h.append(g)           
  return h
```

```
#utils
from statistics import mean
def avgg(c):
  kj=[]
  r=[]
  for i in c:
    i=i.tolist()
    r.append(i)
  for i in r:
    for j in i:
      for k in j:
        kj.append(k)
  return mean(kj)      
```

```
#init pop and sort by cost
import random as rd
import numpy as np
from statistics import mean
n=10
k=0
pop=[]
avg=[]
gen=6
for i in range(gen):
    for i in range(n):
      while len(pop)<10:
        a=[]
        for i in range(6):
          a.append(round(rd.uniform(0, 39), 2))
          a.append(round(rd.uniform(0, 49), 2))
        if overlap(a)==True:   
          pop.append(a)
        k=k+1  
    pop=numpy.array(pop)
    pop.shape


    cost=[]
    for i in pop:
      i=i.reshape((1,12))
      ol=model.predict(i)
      cost.append(ol)
    #print(cost)
    avg.append(avgg(cost))
    pop=pop.tolist()
    #print(pop)
    zipped_lists = zip(cost, pop)
    sorted_pairs = sorted(zipped_lists)

    tuples = zip(*sorted_pairs)
    cost, pop = [ list(tuple) for tuple in  tuples]
    #print(cost)
    #print(pop)
    #natural selection
    elite_sol=cost[0]
    elite_placement=pop[0]
    print(elite_sol)
    print(elite_placement)
    elite=pop[:6]
    #print(elite)
    elite_costs=cost[:6]
    #print(elite_costs)
    elite_costs=weighter(elite_costs)
    #print(elite_costs)
    elite_costs=prob(elite_costs)
    #print(elite_costs)
    parents=rd.choices(elite, weights=elite_costs, k=4) 
    #print(parents) 
    children=[] 
    for u in range(0,len(parents),2):
      beta=rd.uniform(0,1)
      offspring1=[]
      offspring2=[]
      for i in range(12):
        offspring1.append(beta*parents[u][i]+(1-beta)*parents[u+1][i])
        offspring2.append((1-beta)*parents[u][i]+beta*parents[u+1][i])
      children.append(offspring1) 
      children.append(offspring2)
    pop=elite+children   
    #print(pop)
    pop=mutation(pop,0.2)
    #print(pop)
print(avg)    
```
